```python
import os
import pandas as pd

cell_path = r'E:\YuCheng\cell_data_processed\cells'
radar_path = r'E:\YuCheng\cell_data_processed\radar\grids\global'

# ËíêÈõÜ‰∏ç‰∏ÄËá¥ÁöÑÁµêÊûú
inconsistent_files = []

# ÂàóÂá∫ÊâÄÊúâ csv
for csv_file in os.listdir(cell_path):
    if not csv_file.endswith('.csv'):
        continue

    csv_name = os.path.splitext(csv_file)[0]
    csv_path = os.path.join(cell_path, csv_file)
    radar_folder = os.path.join(radar_path, csv_name)

    # Ê™¢Êü•Â∞çÊáâÁöÑÈõ∑ÈÅîË≥áÊñôÂ§æÊòØÂê¶Â≠òÂú®
    if not os.path.isdir(radar_folder):
        # print(f'‚ö†Ô∏è Áº∫Â∞ëÈõ∑ÈÅîË≥áÊñôÂ§æ: {csv_name}')
        inconsistent_files.append((csv_name, 'folder_not_found'))
        continue

    # ËÆÄÂèñ CSVÔºåË®àÁÆóË≥áÊñôÁ≠ÜÊï∏
    try:
        df = pd.read_csv(csv_path)
        csv_len = len(df)
    except Exception as e:
        print(f'‚ùå ÁÑ°Ê≥ïËÆÄÂèñ CSV: {csv_name}, error: {e}')
        continue

    # Áµ±Ë®àÂúñÊ™îÊï∏
    radar_files = [f for f in os.listdir(radar_folder) if f.endswith('.npy')]
    radar_len = len(radar_files)

    if csv_len != radar_len:
        print(f'‚ùå ‰∏ç‰∏ÄËá¥: {csv_name} ‚Üí CSV Èï∑Â∫¶ = {csv_len}, ÂúñÊï∏ = {radar_len}')
        inconsistent_files.append((csv_name, csv_len, radar_len))

# Áµ±Ë®à
print("\n=== ‰∏ç‰∏ÄËá¥Á∏ΩÁµê ===")
print(f"ÂÖ±ÁôºÁèæ {len(inconsistent_files)} Á≠Ü‰∏ç‰∏ÄËá¥ÁöÑË≥áÊñôÔºö")
for item in inconsistent_files:
    print(item)
```
