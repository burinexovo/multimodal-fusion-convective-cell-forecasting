```python
import json

with open('ffnn_haversine_errors.json', 'r', encoding='utf-8') as f:
    ffnn_haversine_errors = json.load(f)

with open('lstm_haversine_errors.json', 'r', encoding='utf-8') as f:
    lstm_haversine_errors = json.load(f)
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

jheng_hei_b_font = fm.FontProperties(fname=r'fonts\微軟正黑體_粗體.ttc')
jheng_hei_font = fm.FontProperties(fname=r'fonts\微軟正黑體_標準.ttc')

plt.figure(figsize=(6, 5))
plt.boxplot([ffnn_haversine_errors, lstm_haversine_errors])
# plt.title("FFNN Haversine Distance Error Distribution")
plt.ylabel("Haversine 距離誤差 (km)", fontproperties=jheng_hei_b_font, fontsize=12)
plt.xticks([1, 2], ['FFNN 預測誤差', 'LSTM 預測誤差'], fontproperties=jheng_hei_b_font)
plt.grid(True, axis='y', linewidth=0.5, color='gray', linestyle='--', alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

plt.hist(ffnn_haversine_errors, bins=40, alpha=0.5, label='FFNN 誤差分布', color='#4C72B0', edgecolor='black')
plt.hist(lstm_haversine_errors, bins=40, alpha=0.5, label='LSTM 誤差分布', color='#D62728', edgecolor='black')

plt.xlabel('Haversine 距離誤差 (km)', fontproperties=jheng_hei_b_font, fontsize=12)
plt.ylabel('頻率', fontproperties=jheng_hei_b_font, fontsize=12)
plt.legend(prop=jheng_hei_font, fontsize=10)
plt.grid(True, linewidth=0.5, color='gray', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
