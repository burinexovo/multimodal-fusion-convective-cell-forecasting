```python
import calendar
# 10、11、12月

for month in [10, 11, 12]:
    print(f'{month}月的天數：{calendar.monthrange(2022, month)[1]}')
    days_in_month = calendar.monthrange(2022, month)[1]
    part1_end = int(days_in_month * 0.7)
    part2_end = int(days_in_month * 0.85)
    print(f'訓練：1 ~ {part1_end}')
    print(f'驗證：{part1_end + 1} ~ {part2_end}')
    print(f'測試：{part2_end + 1} ~ {days_in_month}')
```
